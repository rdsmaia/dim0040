# bibliotecas

```python
# carrega bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

# criação da variável x

```python
# inicialização de x
x = np.arange(-5,5,0.1)
print(f'A variável x= {x}')
```

# determinação de y = f(x)

```python
# número de elementos de x
nelem = len(x)

# inicializa y com zeros e mesmo tamanho de x
y = np.zeros(np.shape(x))

# gera y
for i in range(nelem):
    
    # intervalo mais à esquerda
    if x[i] <= -1:
        y[i] = x[i] + 1
    
    # meio
    elif x[i] > -1 and x[i] <= 2:
        y[i] = -x[i]**2 + 2
    
    # intervalo mais à direita
    else:
        y[i] = x[i]
print(f'y ={y}')
```

# gráfico

```python
# primeira parte de y
y1 = [y[i] for i in range(nelem) if x[i]<=-1]
x1 = [x[i] for i in range(nelem) if x[i]<=-1]

# segunda parte de y
y2 = [y[i] for i in range(nelem) if x[i]>-1 and x[i]<=2]
x2 = [x[i] for i in range(nelem) if x[i]>-1 and x[i]<=2]

# terceira parte de y
y3 = [y[i] for i in range(nelem) if x[i]>2]
x3 = [x[i] for i in range(nelem) if x[i]>2]

# mostra gráfico
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.plot(x3,y3)

plt.show()
```

# importação de bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

# cria variável x

```python
x = np.arange(-10,10)
print(f'Aqui está o vetor criado: x={x}')
```

# define a função f(x)

```python
y1 = np.cbrt(x) + x**3 + 7
print(f'y={y1}')
```

# produz o gráfico de f(x)

```python
plt.plot(x,y1)
plt.show()
```
